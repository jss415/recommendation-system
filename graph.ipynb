```python
RMSEs = [1.01, 0.96, 0.98, 0.94, 1.08, 0.87]
MAEs = [0.82, 0.76, 0.79, 0.73, 0.85, 0.69]
```

```python
precisions = [0.89, 0.92, 0.90, 0.94, 0.82, 0.88]
recalls = [0.56, 0.60, 0.58, 0.64, 0.53, 0.65]
F_measures = [0.69, 0.72, 0.70, 0.77, 0.64, 0.66]
NDCGs = [0.96, 0.97, 0.97, 0.99, 0.94, 0.99]
```

```python
x_labels = ['user_user', 'item_item', 'KNN', 'neural', 'content', 'surprise_SVD']
```

```python
import matplotlib.pyplot as plt
```

```python
# create a figure and axis object
fig, ax = plt.subplots()

# plot the RMSE line with x-labels
ax.plot(x_labels, RMSEs, label='RMSE')

# plot the MAE line with x-labels
ax.plot(x_labels, MAEs, label='MAE')

# set the title and axis labels
ax.set_title('RMSE and MAE Comparison')
ax.set_xlabel('Model Type')
ax.set_ylabel('Error Value')

# add a legend
ax.legend()

# show the plot
plt.show()
```

```python
# create a figure and axis object
fig, ax = plt.subplots()

# plot the Precision line with x-labels
ax.plot(x_labels, precisions, label='Precision')

# plot the Recall line with x-labels
ax.plot(x_labels, recalls, label='Recall')

# plot the F-measure line with x-labels
ax.plot(x_labels, F_measures, label='F-measure')

# plot the NDCG line with x-labels
ax.plot(x_labels, NDCGs, label='NDCG')

# set the title and axis labels
ax.set_title('Precision, Recall, F-measure, and NDCG Comparison')
ax.set_xlabel('Model Type')
ax.set_ylabel('Metric Value')

# add a legend
ax.legend()

# show the plot
plt.show()
```
